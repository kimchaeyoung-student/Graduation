```
import folium
from pandas import DataFrame   # 데이터 분석 패키지
from pandas import ExcelFile
```

```
filename4 = '/content/drive/My Drive/onlydental.csv'
onlydental = pd.read_csv(filename4)
```

```
onlydental
```

```
onlydental2 = onlydental.dropna(axis=0)
# nan 데이터 제거
```

```
onlydental2
```

```
seoul = onlydental2.loc[onlydental2['도로명전체주소'].str.contains('서울특별시',na=False)]
seoul
```

```
map_osm = folium.Map(location=[37.566647,126.978426], zoom_start=10, tiles='Stamen Terrain')

for i in seoul.index:
  name = seoul.loc[i, '업태구분명']
  lat = seoul.loc[i,'위도']
  lng = seoul.loc[i,'경도']

  marker = folium.Marker([lat,lng], popup=name)
  marker.add_to(map_osm)

map_osm.save('서울의_치과병원분포.html')
map_osm
```

```
map_osm = folium.Map(location=[37.566647,126.978426], zoom_start=10, tiles='Stamen Terrain')

for i in subway.index:
    # 행 우선 접근 방식으로 값 추출하기
    name = subway.loc[i, '지하철역명']
    lat = subway.loc[i, '위도']
    lng = subway.loc[i, '경도']
    
    # 추출한 정보를 지도에 표시
    folium.CircleMarker([lat,lng],
                        radius = 5,
                        color = 'blue',
                        fill = True,
                        rapopup=name).add_to(map_osm)

map_osm.save('서울의_지하철역_분포.html')
map_osm
```
