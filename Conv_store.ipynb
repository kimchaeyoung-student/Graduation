```
import folium
from pandas import DataFrame   # 데이터 분석 패키지
from pandas import ExcelFile
```

```
#불러올 파일의 경로를 filename 변수에 저장
filename2 = '/content/drive/My Drive/Conv_store_s.csv'
```

```
Conv_store = pd.read_csv(filename2)
Conv_store
```

```
Conv_store = Conv_store.filter(['MEGA_NM','X_Lon_경도','Y_Lat_위도'])
Conv_store
```

```
map_osm = folium.Map(location=[37.566647,126.978426], zoom_start=20, tiles='http://api.vworld.kr/req/wmts/1.0.0/D05C77C9-AB62-3E70-9183-0E044A461BBD/gray/{z}/{y}/{x}.png',
    attr='YworldGray')

for i in Conv_store.index:
  name = Conv_store.loc[i, 'MEGA_NM']
  lat = Conv_store.loc[i,'Y_Lat_위도']
  lng = Conv_store.loc[i,'X_Lon_경도']

  folium.CircleMarker([lat,lng],
                      radius = 5,
                      color = 'green',
                      fill = True,
                      rapopup = name).add_to(map_osm)

map_osm.save('서울의_편의점_분포.html')
map_osm
```
